#Basic problem

This is an example of an "Exactly one" problem.

Binary variables: $r, g, b$
Each will equal 1 if the region has that color; 0 otherwise.

Write a function which is minimized when exactly one of ${r,g,b}$ is turned on.


The QUBO: $(1-r-g-b) ^ 2$

Why squared? 


Reduced QUBO: $1-r-g-b+2br+2gr+2bg$

Expected solutions:

{b=0;g=0;r=1} function value= -1.0

{b=0;g=1;r=0} function value= -1.0

{b=1;g=0;r=0} function value= -1.0

{b=0;g=0;r=0} function value= 0.0

{b=0;g=1;r=1} function value= 0.0

{b=1;g=1;r=0} function value= 0.0

{b=1;g=0;r=1} function value= 0.0

{b=1;g=1;r=1} function value= 3.0

States with function values = -1.0 are Ground states, while other states are Excited states

## Develop Ising or QUBO model

1. Start with a problem in the real world
2. Write a minimization function that incorporates:
- Constraints: The rules that we have to follow
- Objective: What we want to minimize

General form:
Ising or QUBO = $min(objective + y(constraints))$


### 1. Write Objective and constraints in problem domain

Basic problem example(Using QUBO process):

#### Problem: Pick the pair of boxes with the smallest sum.

Box 1 : 17 

Box 2 : 21 

Box 3 : 19

Objective: Minimize the sum of the boxes chosen

Constraints: Pick exactly two boxes

##### Convert objective and constraints to math statements with binary variables

$x_{i} = 1$ if box $i$ is selected, 0 otherwise

Objective:
$min(17x_{1} + 21x_{2} + 19x_{3})$

Constraints:
$x_{1} + x_{2} + x_{3} = 2$ (Pick two boxes)

##### Make objective and constraints "QUBO appropriate"

QUBO:
$min(17x_{1} + 21x_{2} + 19x_{3}) + \gamma(x_{1}+x_{2}+x_{3}-2)^{2}$

### Graph Partitioning

Given a graph G, split the graph into two subsets such that:

1. The subsets are the same size.
2. There are as few edges as possible between the two subsets.

Objective:
- Minimize the number of edges between subsets

Constraints:
- The two subsets have equal size

We use binary variables to "label" our nodes:

$x_{i} =$

1 if node i is in set A 

0 if node i is in set B

Objective:

Minimize the number of edges between subsets.
